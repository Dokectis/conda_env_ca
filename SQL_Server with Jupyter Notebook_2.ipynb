```python

import pyodbc
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Define your connection parameters
server = 'SQLSERVERHOST'
database = 'sandbox'

# Create an engine
engine = create_engine(f'mssql+pyodbc://{server}/{database}?driver=ODBC+Driver+18+for+SQL+Server&TrustServerCertificate=yes')
```

```python
# Write a query
query = "SELECT * FROM customer"

# Execute the query and fetch data into a DataFrame
df = pd.read_sql(query, engine)

# Display the DataFrame
print(df.head())
```

```python
df
```

```python
# Fetch data from the first table
query1 = "SELECT * FROM table1"
df1 = pd.read_sql(query1, engine)

# Fetch data from the second table
query2 = "SELECT * FROM table2"
df2 = pd.read_sql(query2, engine)

# Fetch data from the third table
query3 = "SELECT * FROM table3"
df3 = pd.read_sql(query3, engine)

# Display the DataFrames
print("Data from table1:")
print(df1.head())

print("\nData from table2:")
print(df2.head())

print("\nData from table3:")
print(df3.head())

# Example: Merging data from two tables
merged_df = pd.merge(df1, df2, on='common_column')

# Display the merged DataFrame
print("Merged DataFrame:")
print(merged_df.head())
```

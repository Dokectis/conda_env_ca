```python
# Import necessary packages
from sqlalchemy import create_engine, inspect
import pandas as pd
```

```python
# Define your connection parameters
server = 'SQLSERVERHOST'
database = 'sandbox'

# Create an engine
engine = create_engine(f'mssql+pyodbc://{server}/{database}?driver=ODBC+Driver+18+for+SQL+Server&TrustServerCertificate=yes')
```

```python

# Create an inspector
inspector = inspect(engine)

# Get a list of all table names
tables = inspector.get_table_names()


```

```python
# Display the list of tables
print("Tables in the database:")
for table in tables:
    print(table)
```

```python
# Optionally, fetch data from each table
for table in tables:
    query = f"SELECT * FROM {table}"
    df = pd.read_sql(query, engine)
    print(f"\nData from table {table}:")
    print(df.head())
```

```python
# Display the list of tables
print("Tables in the database:")
for table in tables:
    print(table)

# Dictionary to hold DataFrames
dataframes = {}

```

```python
# Fetch data from each table and store in the dictionary
for table in tables:
    query = f"SELECT * FROM {table}"
    df = pd.read_sql(query, engine)
    dataframes[table] = df
    print(f"\nData from table {table}:")
    print(df.head())


```

```python
# Example: Access a specific DataFrame by table name
table_name = 'product_order'  # Replace with your table name
df = dataframes[table_name]
```
